```python
def exp(lam):
  x = float((-np.log(np.random.uniform()))/lam)
  return x
```

```python
def density(x,lam=1):
  return lam*np.exp(-lam*x)
```

```python
exp(5)
```

```python
def mult_exp(lam,n):
  vet = []
  for i in range(n):
   vet.append(exp(lam))
  return vet
```

```python
mult_exp(3,1000000)
```

```python
sample = mult_exp(1,1000)

x = np.linspace(0,5,1000)
plt.plot(x, density(x))
plt.hist(sample, density=True)

```

```python
def geo_inv(p):
  u = np.random.rand()
  i = (np.log(u)/np.log(1-p)) + 1
  return i
```

```python
def geo_exp(lam, n):
  p = lam/n
  x = geo_inv(p)/n
  return x
```

```python
geo_exp(0.5,1)
```

```python
def mult_exp_geo(lam,n,m):
  vet = []
  for i in range(m):
    vet.append(geo_exp(lam,n))
  return vet
```

```python
plt.hist(mult_exp_geo(0.5,1,1000))
```

```python
def poisson(lam):
    u = np.random.rand()
    i = 0
    p0 = np.exp(-lam)
    F = p0
    while u > F:
      i = i + 1
      p0 = (lam/i)*p0
      F = F + p0
    return i
```

```python
def poisson_via_exp(mu):
  temp = 0
  i = 0
  while temp <= 1:
    temp += exp(mu)
    i = i + 1
  return i - 1
```

```python
poisson_via_exp(1)
```

```python
def poisson_via_exp_mult(mu,n):
  vet = []
  for i in range(n):
    vet.append(poisson_via_exp(mu))
  return vet
```

```python
poiss_vet = poisson_via_exp_mult(4,100000)
poiss_vet
```
