```python
import numpy as np
import matplotlib.pyplot as plt
import time

def bernoulli(p):
  u = np.random.rand()
  if u <= p:
    return 1
  else:
    return 0
```

```python
prob = 0.5
tam = 10

def gera_binomial_rec(tam, prob):
    k = 0
    pi = (1 - prob) ** tam
    F = pi
    u = np.random.uniform(0, 1)
    passos = 0

    while u > F:
        fator = ((tam - k) / (k + 1)) * (prob / (1 - prob))
        pi *= fator
        k += 1
        F += pi
        passos += 1
    return k, passos
```

```python
tam = 20
num_sims = 100000
lista_p = [0.1, 0.3, 0.5, 0.7, 0.9]

soma_p1, soma_p2, soma_p3, soma_p4, soma_p5 = [], [], [], [], []
rec_p1, rec_p2, rec_p3, rec_p4, rec_p5 = [], [], [], [], []
steps_p1, steps_p2, steps_p3, steps_p4, steps_p5 = [], [], [], [], []

tempos_soma = []
tempos_rec = []
```

```python
#p = 0.1
inicio = time.time()
for i in range(num_sims):
    total = 0
    for j in range(tam):
        total += bernoulli(lista_p[0])
    soma_p1.append(total)
fim = time.time()
tempos_soma.append(fim - inicio)
print(fim - inicio)

inicio = time.time()
for i in range(num_sims):
    valor, passos = gera_binomial_rec(tam, lista_p[0])
    rec_p1.append(valor)
    steps_p1.append(passos)
fim = time.time()
tempos_rec.append(fim - inicio)
print(fim - inicio)
print(np.mean(steps_p1))
```
