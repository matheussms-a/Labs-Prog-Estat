ExercÃ­cio 1

```python
def exp(lam):
  x = float((-np.log(np.random.uniform()))/lam)
  return x
```

```python
def norm_abs():

  y = exp(1)
  u = np.random.uniform()
  while u > np.exp(-((y-1)**2)/2):
    y = exp(1)
    u = np.random.uniform()
  return y
```

```python
def norm_abs_mult(n):
  lista = []
  for i in range(n):
    lista.append(norm_abs())
  return lista
```

```python
amostra_abs = norm_abs_mult(100000)
```

```python
plt.hist(amostra_abs)
```

```python
def bernoulli(p):
  u = np.random.rand()
  if u <= p:
    return 1
  else:
    return 0
```

```python
def norm_padrao_(n):
  lista = []
  for i in range(n):
    if bernoulli(0.5) == 1:
       lista.append(norm_abs())
    else: lista.append(-norm_abs())
  return lista
```

```python
nnorm = norm_padrao_(100000)
plt.hist(nnorm, bins = 30, density=True)
```

```python
def norm_ger(mu, sigma2, n):
  lista = []
  nnorm = norm_padrao_(n)
  for i in range (n):
    lista.append(np.sqrt(sigma2)*(nnorm[i]+mu))
  return lista
```

```python
plt.hist(norm_ger(10,1,100000))
```

ExercÃ­cio 2

```python
def distr():
  y = np.random.uniform(0,np.pi)
  u = np.random.uniform()
  while u > (math.sin(y)):
    y = np.random.uniform(0,np.pi)
    u = np.random.uniform()
  return y
```

```python
def distr_mult(n):
  lista = []
  for i in range(n):
    lista.append(distr())
  return lista
```

```python
plt.hist(distr_mult(100000), bins=50, density=True)
```
