```python
import numpy as np
```

```python
# ex 1 np
vetor = np.random.randint(0, 101, 20)
print(vetor)
print(np.mean(vetor))
print(np.std(vetor))
print(np.max(vetor))
```

```python
# ex 2 np
matriz = np.random.randn(5, 5)
print("Matriz:\n", matriz)
print("Soma das linhas:", np.sum(matriz, axis=1))
print("Soma das colunas:", np.sum(matriz, axis=0))
print("Diagonal principal:", np.diag(matriz))
```

```python
# ex 3 np
def analisa_vetores(v1, v2):
    produto_escalar = np.dot(v1, v2)
    distancia = np.linalg.norm(v1 - v2)
    return produto_escalar, distancia

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print(analisa_vetores(v1, v2))
```

```python
# ex 4 np
A = np.random.randint(-5, 6, (4, 4))
print("Matriz A:\n", A)

det = np.linalg.det(A)
print("Determinante:", det)

if det != 0:
    A_inv = np.linalg.inv(A)
    print("Inversa de A:\n", A_inv)

    identidade = np.dot(A, A_inv)
    print("A * A_inv:\n", identidade)
else:
    print("A não é invertível")
```

```python
# ex 5 np
M = np.random.rand(5, 5)
B = 0.5 * (M + M.T)
print("Matriz simétrica B:\n", B)

valores, vetores = np.linalg.eig(B)
print("Autovalores:", valores)
print("Autovetores:\n", vetores)

for i in range(len(valores)):
    lado_esq = np.dot(B, vetores[:, i])
    lado_dir = valores[i] * vetores[:, i]
    print("Diferença vetor", i, ":", np.allclose(lado_esq, lado_dir))
```

```python
# ex 6 np
A = np.random.randint(1, 10, (3, 3))
y = np.random.randint(1, 10, 3)
print("A:\n", A)
print("y:", y)

x = np.linalg.solve(A, y)
print("Solução x:", x)

erro = np.linalg.norm(np.dot(A, x) - y)
print("Erro:", erro)
```

```python
# ex 7 np
dados = np.random.randn(2, 100)
print("Shape dos dados:", dados.shape)

S = np.cov(dados)
print("Matriz de covariância:\n", S)

valores, vetores = np.linalg.eig(S)
print("Autovalores:", valores)
print("Autovetores:\n", vetores)

for i in range(len(valores)):
    print(f"Direção {i}: {vetores[:, i]} (variância {valores[i]})")
```

```python
import pandas as pd
```

```python
# ex 1 pd
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
print(df.head())
print(df.info())
print(df.isna().sum())
```

```python
#ex 2 pd
estatisticas = df.groupby('species')['petal_length'].agg(['mean', 'std'])
print(estatisticas.sort_values('mean'))
```

```python
# ex 3 pd

dados = {
    "nome": ["Ana", "Bruno", "Carla"],
    "idade": [20, 25, 34],
    "cidade": ["SP", "RJ", "BH"]
}
df2 = pd.DataFrame(dados)

df2["idade_quadrado"] = df2["idade"] ** 2
print(df2)

df2 = df2.drop(columns=["idade_quadrado"])
print(df2)
```

```python
# ex 4 pd
novo_df = pd.DataFrame({
    'nome': ['Ana', 'João', 'Maria'],
    'idade': [25, 30, 35],
    'cidade': ['São Paulo', 'Rio', 'Belo Horizonte']
})

novo_df['idade_quadrado'] = novo_df['idade']**2
novo_df = novo_df.drop('idade_quadrado', axis=1)
print(novo_df)
```

```python
# ex 4 pd

filtro = df2[df2["idade"] > 30]
print(filtro)

print("Qtd de linhas:", filtro.shape[0])
```

```python
# ex 5 pd

notas = pd.DataFrame({
    "Matemática": [7, 8, 9],
    "Português": [6, 9, 8],
    "História": [5, 7, 6]
}, index=["Aluno1", "Aluno2", "Aluno3"])

print(notas)

media_aluno = notas.mean(axis=1)
print("Média por aluno:\n", media_aluno)

media_disc = notas.mean(axis=0)
print("Média por disciplina:\n", media_disc)

print("Aluno com maior média:", media_aluno.idxmax())
```

```python
# ex 6 pd

petalas = df[['petal_length', 'petal_width']]
normalizado = (petalas - petalas.min()) / (petalas.max() - petalas.min())
print(normalizado.head())
```

```python
# ex 7 pd

vendas = pd.DataFrame({
    'produto': ['A', 'B', 'A', 'C'],
    'preço': [10, 20, 10, 30],
    'quantidade': [2, 1, 3, 2]
})

vendas['receita'] = vendas['preço'] * vendas['quantidade']
receita_total = vendas.groupby('produto')['receita'].sum()
print(receita_total.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
```

```python
# ex 1 plt

plt.hist(df['sepal_length'], bins=20)
plt.title('Distribuição do Comprimento da Sépala')
plt.xlabel('Comprimento (cm)')
plt.ylabel('Frequência')
plt.show()
```

```python
# ex 2 plt

cores = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

for especie in df['species'].unique():
    dados = df[df['species'] == especie]
    plt.scatter(dados['sepal_length'], dados['sepal_width'],
                label=especie, alpha=0.7)

plt.legend()
plt.title('Relação entre Comprimento e Largura da Sépala')
plt.xlabel('Comprimento da Sépala (cm)')
plt.ylabel('Largura da Sépala (cm)')
plt.show()
```

```python
# ex 3 plt

plt.scatter(df['petal_length'], df['petal_width'],
            s=df['sepal_length']*10, alpha=0.5)
plt.title('Relação entre Comprimento e Largura da Pétala')
plt.xlabel('Comprimento da Pétala (cm)')
plt.ylabel('Largura da Pétala (cm)')
plt.show()
```

```python
# ex 4 plt

fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, especie in enumerate(df['species'].unique()):
    dados = df[df['species'] == especie]
    axs[i].hist(dados['petal_length'], bins=15)
    axs[i].set_title(especie)
    axs[i].set_xlabel('Comprimento da Pétala (cm)')

plt.tight_layout()
plt.show()
```

```python
# ex 5 plt

colunas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.boxplot([df[col] for col in colunas], labels=colunas)
plt.title('Boxplot das Características das Flores')
plt.ylabel('Centímetros')
plt.show()
```

```python
# ex 6 plt

medias = df.groupby('species')[colunas].mean()
medias.plot(kind='bar')
plt.title('Médias das Características por Espécie')
plt.ylabel('Centímetros')
plt.xticks(rotation=0)
plt.show()
```

```python
# ex 7 plt

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

for especie in df['species'].unique():
    dados = df[df['species'] == especie]
    ax.scatter(dados['sepal_length'],
               dados['sepal_width'],
               dados['petal_length'],
               label=especie)

ax.set_xlabel('Comprimento da Sépala (cm)')
ax.set_ylabel('Largura da Sépala (cm)')
ax.set_zlabel('Comprimento da Pétala (cm)')
ax.legend()
plt.show()
```
