```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def calcular_integral_exp_quad(tamanho_amostra, limite_inf, limite_sup):
    dados_x = np.random.uniform(limite_inf, limite_sup, size=tamanho_amostra)
    valores_y = np.exp(dados_x)
    resultado = np.mean(valores_y) * (limite_sup - limite_inf)
    return resultado

print(calcular_integral_exp_quad(1000, 0, 1))
```

Estimar integral de 0 a 1 de exp{⁻x²}

```python
def estimacao_pi_area(iteracoes):
    x_coords = np.random.uniform(0, 1, iteracoes)
    y_coords = np.random.uniform(0, 1, iteracoes)

    dentro_circulo = (x_coords**2 + y_coords**2) <= 1
    contagem = np.sum(dentro_circulo)

    estimativa = 4 * (contagem / iteracoes)
    return estimativa

print(estimacao_pi_area(5000))
```

```python
def estatisticas_monte_carlo(n, a, b):
    valores = np.random.uniform(a, b, n)
    media_est = np.mean(valores)
    var_est = np.var(valores, ddof=1)
    return media_est, var_est

print(estatisticas_monte_carlo(1000, 0, 1))
```

Estimar esperança e variança de uniformes

```python
def simular_integral_dupla(n):
    u = np.random.uniform(0, 1, n)
    v = np.random.uniform(0, 1, n)
    funcao_avaliada = np.exp(-(u**2 + v**2))
    return np.mean(funcao_avaliada)

print(simular_integral_dupla(1000))

#5
```

integral dupla no quadrado 0,0 0,1 1,0 1,1 de exp{-(x²+y²)}

```python
def integral_exponencial_transformada(qtd):
    u = np.random.uniform(0, 1, qtd)
    x_exp = -np.log(u)
    resultados = np.exp(-x_exp)
    return np.mean(resultados)

print(integral_exponencial_transformada(100))
```

integral de exp{-x}

```python
def integral_racional(n_simulacoes):
    bases = np.random.uniform(0, 1, n_simulacoes)
    variavel_transf = -np.log(bases)
    estimativa = np.mean(1 / (1 + variavel_transf))
    return estimativa

print(integral_racional(5000))
```

integral de exp{-x}/(1+x)

```python
def integral_sinc_monte_carlo(pontos):
    uni = np.random.uniform(0, 1, pontos)
    x_vals = -np.log(uni)
    termos = np.sin(x_vals) / (x_vals * np.exp(-x_vals))
    return np.mean(termos)

print(integral_sinc_monte_carlo(5000))
```
