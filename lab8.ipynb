```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Ex 1

```python
def antit1(n):
  A=[]
  for i in range(n):
    U=np.random.uniform(0,1)
    V=1-U
    A.append((np.exp(U+1)+np.exp(V+1)))
  return np.mean(A)/2
```

```python
def montecarlo1(n):
  A=[]
  for i in range(n):
    U=np.random.uniform(0,1)
    A.append(np.exp(U+1))
  return np.mean(A)
```

```python
AmVI=[]
AmMC=[]
for i in range(10**3):
  AmVI.append(antit1(10**3))
for i in range(10**3):
  AmMC.append(montecarlo1(10**3))
```

```python
plt.hist([AmVI,AmMC],bins=40, color=["green",'red'],label=["Antitética","Montecarlo"])
plt.legend()
#plt.hist(AmMC,bins=20, color="red")
plt.show()
print("variância Variável Antitética:",np.var(AmVI)," ,variância Montecarlo:", np.var(AmMC), "\n razão entre variâncias: ", np.var(AmVI)/np.var(AmMC))
```

```python
def exp_inv(lam):
  U=np.random.uniform()
  x=np.log(U)*(-1/lam)
  return x
```

```python
def antit2(n):
  A=[]
  for i in range(n):
    U=np.random.uniform(0,1)
    V=1-U
    E1=-np.log(U)
    E2=-np.log(V)
    A.append(np.exp(-E1**2)+np.exp(-E2**2))
  return np.mean(A)/2
```

```python
def montecarlo2(n):
  A=[]
  for i in range(n):
    U=np.random.uniform(0,1)
    E1=-np.log(U)
    A.append(np.exp(-E1**2))
  return np.mean(A)
```

```python
AmVI2=[]
AmMC2=[]
for i in range(10**3):
  AmVI2.append(antit2(10**3))
for i in range(10**3):
  AmMC2.append(montecarlo2(10**3))
```

```python
plt.hist([AmVI2,AmMC2],bins=40, color=["green",'red'],label=["Antitética","Montecarlo"])
plt.legend()
#plt.hist(AmMC,bins=20, color="red")
plt.show()
print("variância Variável Antitética:",np.var(AmVI2)," ,variância Montecarlo:", np.var(AmMC2), "\n razão entre variâncias: ", np.var(AmVI2)/np.var(AmMC2))
```

```python
def normal01():
  a=0
  while a==0:
    U=np.random.uniform()
    E=np.log(U)*(-1)
    Y=E
    U2=np.random.uniform()
    if U2<np.exp((-(Y-1)**2)/2):
      a=1
  U3=np.random.uniform()
  if U3<=0.5:
    Y=-Y
  return Y
def f(x):
  return -x**2 + 1
```

```python
def antit3(n):
  A=[]
  for i in range(n):
    Z=normal01()
    V=-Z
    A.append(f(np.sqrt(2)*Z+1)+f(np.sqrt(2)*V+1))
  return np.mean(A)/2
```

```python
def montecarlo3(n):
  A=[]
  for i in range(n):
    N=normal01()
    A.append(f(np.sqrt(2)*N+1))
  return np.mean(A)
```

```python
AmVI3=[]
AmMC3=[]
for i in range(10**3):
  AmVI3.append(antit3(10**3))
for i in range(10**3):
  AmMC3.append(montecarlo3(10**3))
```

```python
plt.hist([AmVI3,AmMC3],bins=40, color=["green",'red'],label=["Antitética","Montecarlo"])
plt.legend()
#plt.hist(AmMC,bins=20, color="red")
plt.show()
print("variância Variável Antitética:",np.var(AmVI3)," ,variância Montecarlo:", np.var(AmMC3), "\n razão entre variâncias: ", np.var(AmVI3)/np.var(AmMC3))
```
